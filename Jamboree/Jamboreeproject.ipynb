```python
#standardizing the data so that the model is not biased towards a specific value
jamboree_data_stand = jamboree_data.copy()
for col in jamboree_data.columns:
    jamboree_data_stand[col]=(jamboree_data_stand[col]-jamboree_data_stand[col].mean())/jamboree_data_stand[col].std()
jamboree_data_stand
```

```python
sns.scatterplot(x='GRE Score',y='Chance of Admit ',data=jamboree_data_stand)
plt.show()
```

## Splitting data into test and train

```python
x = jamboree_data_stand.drop('Chance of Admit ',axis=1)
y = jamboree_data_stand['Chance of Admit ']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7 , random_state=100)
```

```python
x_train.shape
```

```python
y_train.shape
```

### Linear Regression

```python
#importing the library required for building a linear regression model
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
#fit model to training data
model.fit(x_train,y_train)
```

### Evaluating the model

```python
model.intercept_
```

```python
#Creating a dataframe for visualizing the coefficients of the model.This will help us find out the feature that is given more weight by the model.
coeff_df = pd.DataFrame(model.coef_,x_test.columns,columns=['Coefficient'])
coeff_df
```

```python
#Passsing the test dataset values to model for it to predict the chance of admission.This will help us validate the model.
y_pred = model.predict(x_test)
```

```python
#Passing the train dataset values to model for it to predict the chance of aadmission. 
y_pred_train  = model.predict(x_train)
```

```python
#Evaluating the model performance using the mean squared error and r-squared error metrics.
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error_train = mean_squared_error(y_train,y_pred_train)
mean_squared_error_test = mean_squared_error(y_test,y_pred)
r2_test = r2_score(y_test,y_pred)
r2_train = r2_score(y_train,y_pred_train)
print("R-squared value for test dataset:",r2_test)
print("R-squared value for train dataset:",r2_train)
print("mean_squared_error_test value for test dataset:",mean_squared_error_test)
print("mean_squared_error_train value for train dataset:",mean_squared_error_train)
```

__We cann see that the model is a good model as the R2 value is 0.8355. This says that 83% of variation in 'chance of admit' is explained by the predictors.__

## OLS Regression

```python
#OLS regression gives a detailed summary of the model.
import statsmodels.api as sm
X_train_sm = x_train

X_train_sm = sm.add_constant(X_train_sm)
lm_1 = sm.OLS(y_train,X_train_sm).fit()

# print the coefficients
lm_1.params
```

```python
print(lm_1.summary())
```

We can see here that the SOP, University rating are insignifcant as the p value is greater than 0.05(If p value is greater than 0.05 we cannot reject the null hypothesis that the coefficient value is equal to 0).This shows that the GRE score is insignificant. But practically gre score is quite significant. The results above might be due to multicollinearity. We will work on multicollinearity and then again try to predict the model.

```python
y_pred_ols = lm_1.predict()
```

## VIF

```python
#VIF measures how much the behaviour of an independent variable is influenced or inflated by its correlation with another variable.
#VIF above 5 indicates that it is highly influenced by the correlation.
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = x_train.columns
vif_data["VIF"] = [variance_inflation_factor(x_train.values, i)
                          for i in range(len(x_train.columns))]
vif_data
```

We see that the VIF for GRE score is above 4. We will do some trial and error to reduce the VIF score.

## HOMOSCEDACITY

Linear regression assumes that the variance is equal along all the values of dependent variable.

```python
residuals = lm_1.resid
data = pd.DataFrame()
data['Predicted'] = y_pred_ols
data['Residuals'] = residuals
sns.lmplot(x='Predicted',y='Residuals',data=data)
```

Based on the above plot we can see that the homoscedacity test is passed.

## Normality of residuals
